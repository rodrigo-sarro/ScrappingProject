## Function opens the website with the informations env

```python
def filmes():

    chrome_path = ('C:/Users/rodri/OneDrive/Área de Trabalho/chromedriver')
    driver = webdriver.Chrome(executable_path=chrome_path)
    
    url = ('https://www.netflix.com/br/login?nextpage=https%3A%2F%2Fwww.netflix.com%2Fbrowse')
    driver.get(url)
    
    load_dotenv('login.env')
    login = os.getenv('user')
    password = os.getenv('password')
    
    elem = driver.find_element_by_xpath('//*[@id="id_userLoginId"]')
    elem.send_keys(login)
    time.sleep(1)
    
    elem = driver.find_element_by_xpath('//*[@id="id_password"]')
    elem.send_keys(password)
    
    enter = driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div/div/div[1]/form/button')
    enter.click()
    time.sleep(2)
    
    perfil = driver.find_element_by_xpath('//*[@id="appMountPoint"]/div/div/div[1]/div[1]/div[2]/div/div/ul/li[2]/div/a/span')
    perfil.click()
    time.sleep(2)
    
    navegar = driver.find_element_by_xpath('//*[@id="appMountPoint"]/div/div/div[1]/div[1]/div[1]/div/div/ul/li[4]/a')
    navegar.click()
    time.sleep(2)
    
    genero = driver.find_element_by_xpath('//*[@id="appMountPoint"]/div/div/div[1]/div[1]/div[1]/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[1]')
    genero.click()
    time.sleep(2)
    
    acao = driver.find_element_by_xpath('//*[@id="appMountPoint"]/div/div/div[1]/div[1]/div[1]/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/ul[1]/li[1]/a')
    acao.click()
    time.sleep(5)
    
    rank = driver.find_element_by_xpath('//*[@id="appMountPoint"]/div/div/div[1]/div[1]/div[1]/div/div[2]/div/div/div/div[2]/div/div/button')
    rank.click()
    time.sleep(1)

#def scraping_filme()
   #response = requests.get('https://www.netflix.com/browse/genre/34399')
```
